```python
import random
from collections import Counter

def throw(N):
    position = 1  
    while position < N:
        coin = random.choice([1, 2])  
        position = position + coin    
    return position 
    

#Above defines one process successfully, either reach or over the aiming landing number
N = int(input('Please input the number of landing on: '))
n = int(input('Please input the times that you want to test:'))

def repeat(n):
    result=[]
    for i in range (n):
        result.append(throw(N))
    return result
#Above repeat the processes until my aiming times reached.

count=Counter(repeat(n))
print(count)
probability=count[N]/n
print('The simulating probability is:', probability)
    
```
